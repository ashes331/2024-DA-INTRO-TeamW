### 월별 미세먼지양

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 파일 열기
f = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {'dates': [], 'pm_values': []} for year in range(2019, 2025)}

# 데이터 처리
for row in data:
    # 첫 번째 열 (날짜)을 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 2019년부터 2024년까지의 데이터만 필터링
    if 2019 <= date.year <= 2024:
        # 해당 연도의 데이터에 추가
        yearly_data[date.year]['dates'].append(date)
        yearly_data[date.year]['pm_values'].append(float(row[1]))

# 파일 닫기
f.close()

# 그래프 그리기
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))  # 2x3 그래프 배열 생성
axes = axes.flatten()  # 2D 배열을 1D로 변환하여 쉽게 접근

for i, year in enumerate(range(2019, 2025)):
    # 각 연도의 데이터를 그리기
    axes[i].plot(yearly_data[year]['dates'], yearly_data[year]['pm_values'])
    axes[i].set_title(f'{year}')  # 연도별
    axes[i].set_xlabel('Month')
    axes[i].set_ylabel('PM')
    axes[i].grid(True)

# 레이아웃 조정
plt.tight_layout()
plt.show()
```

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 파일 열기
f = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 월별 데이터를 저장할 딕셔너리 생성
monthly_data = {month: [] for month in range(1, 13)}  # 1월부터 12월까지 빈 리스트 생성

# 데이터 처리
for row in data:
    # 첫 번째 열 (날짜)을 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 2019년부터 2024년까지의 데이터만 필터링
    if 2019 <= date.year <= 2024:
        # 해당 월에 맞는 미세먼지 값 추가
        month = date.month  # 월 추출
        monthly_data[month].append(float(row[1]))

# 파일 닫기
f.close()

# 월별 평균 미세먼지 값 계산
monthly_avg_pm = [np.mean(monthly_data[month]) if monthly_data[month] else 0 for month in range(1, 13)]

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 월별 평균 미세먼지 값 그래프 그리기
plt.plot(range(1, 13), monthly_avg_pm, marker='o', label='Avg PM')

# 그래프 제목과 레이블 설정
plt.title('Average Monthly PM Data (2019-2024)')
plt.xlabel('Month')
plt.ylabel('PM')
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 파일 열기
f = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별 월별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {month: [] for month in range(1, 13)} for year in range(2019, 2025)}

# 데이터 처리
for row in data:
    # 첫 번째 열 (날짜)을 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 2019년부터 2024년까지의 데이터만 필터링, 2024년은 11월까지만
    if 2019 <= date.year <= 2024:
        if date.year == 2024 and date.month > 11:
            continue  # 2024년 12월 이후는 제외
        # 해당 연도와 월에 맞는 미세먼지 값 추가
        month = date.month
        yearly_data[date.year][month].append(float(row[1]))

# 파일 닫기
f.close()

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 연도별 평균 미세먼지 값 계산 후 그래프 그리기
for year in range(2019, 2025):
    # 각 월별 평균 미세먼지 값 계산 (2024년은 11월까지만)
    monthly_avg_pm = [np.mean(yearly_data[year][month]) if yearly_data[year][month] else 0 for month in range(1, 13)]
    
    # 2024년은 11월까지만 표시
    if year == 2024:
        monthly_avg_pm = monthly_avg_pm[:11]
    
    # 그래프에 그리기
    plt.plot(range(1, len(monthly_avg_pm) + 1), monthly_avg_pm, marker='o', label=f'{year}')

# 그래프 제목과 레이블 설정
plt.title('Average Monthly PM Data by Year (2019-2024)')
plt.xlabel('Month')
plt.ylabel('PM')
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 레이아웃 자동 조정
plt.tight_layout()
plt.legend(title='Year')

# 그래프 출력
plt.show()
```

## 전력사용량

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 파일 열기
f = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

x_result = []
y_result = []
for row in data:
    # 쉼표를 제거하고 전력 사용량을 float으로 변환 후, MW로 변환
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    
    # 결과 리스트에 추가
    y_result.append(power_usage_mw)
    x_result.append(datetime.datetime.strptime(row[0], '%b-%y'))

# 파일 닫기
f.close()

# 그래프 그리기
plt.plot(x_result, y_result)
plt.xlabel('Year-Month')
plt.ylabel('Power Usage (MW)')
plt.grid(True)
plt.show()
```

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 파일 열기
f = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별, 월별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {month: [] for month in range(1, 13)} for year in range(2020, 2025)}

# 데이터 처리
for row in data:
    # 쉼표를 제거하고 전력 사용량을 float으로 변환 후, MW로 변환
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    
    # 날짜를 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 연도와 월에 따라 전력 사용량을 저장
    year = date.year
    month = date.month
    yearly_data[year][month].append(power_usage_mw)

# 파일 닫기
f.close()

# 월별 평균 전력 사용량 계산 (연도별로)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# 연도별 평균 전력 사용량을 계산
average_usage = []

# 각 월에 대한 평균 전력 사용량 계산 (2024년은 8월까지만 포함)
for month in range(1, 13):
    monthly_values = []
    for year in range(2020, 2025):
        # 2024년은 8월까지만 포함
        if year == 2024 and month > 8:
            continue
        # 해당 월의 평균값 계산
        if yearly_data[year][month]:
            monthly_values.append(np.mean(yearly_data[year][month]))
    if monthly_values:
        average_usage.append(np.mean(monthly_values))
    else:
        average_usage.append(0)

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 평균 전력 사용량 그래프 그리기
plt.plot(months[:len(average_usage)], average_usage, label='Average Power Usage', marker='o', color='b')

# 그래프 제목과 레이블 설정
plt.title('Average Monthly Power Usage (MW) from 2020 to 2024 (2024 only until Aug)')
plt.xlabel('Month')
plt.ylabel('Power Usage (MW)')
plt.grid(True)
plt.legend(title='Average')

# X축 레이블 회전
plt.xticks(rotation=45)

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 파일 열기
f = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별, 월별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {month: [] for month in range(1, 13)} for year in range(2020, 2025)}

# 데이터 처리
for row in data:
    # 쉼표를 제거하고 전력 사용량을 float으로 변환 후, MW로 변환
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    
    # 날짜를 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 연도와 월에 따라 전력 사용량을 저장
    year = date.year
    month = date.month
    yearly_data[year][month].append(power_usage_mw)

# 파일 닫기
f.close()

# 월별 평균 전력 사용량 계산 (연도별로)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 각 연도별 데이터 그리기
colors = ['b', 'g', 'r', 'c', 'm']  # 각 연도별로 사용할 색상 리스트
for i, year in enumerate(range(2020, 2025)):
    # 2024년의 경우 8월까지만 표시
    if year == 2024:
        months_to_plot = range(1, 9)  # 2024년은 1월부터 8월까지
    else:
        months_to_plot = range(1, 13)  # 다른 연도는 1월부터 12월까지
    
    # 각 월별 평균 전력 사용량 계산
    avg_monthly_usage = [
        sum(yearly_data[year][month]) / len(yearly_data[year][month]) if len(yearly_data[year][month]) > 0 else 0 
        for month in months_to_plot
    ]
    
    # 연도별 그래프 그리기
    plt.plot([months[month - 1] for month in months_to_plot], avg_monthly_usage, label=f'{year}', color=colors[i], marker='o')

# 그래프 제목과 레이블 설정
plt.title('Monthly Power Usage (MW) from 2020 to 2024 (2024 only until Aug)')
plt.xlabel('Month')
plt.ylabel('Power Usage (MW)')
plt.grid(True)
plt.legend(title='Year')

# X축 레이블 회전
plt.xticks(rotation=45)

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

## 전력사용량과 미세먼지의 관계

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 전력사용량 데이터 처리
f_power = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data_power = csv.reader(f_power)

header_power = next(data_power)

# 연도별, 월별 데이터를 저장할 딕셔너리 생성
yearly_data_power = {year: {month: [] for month in range(1, 13)} for year in range(2020, 2025)}

# 데이터 처리
for row in data_power:
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    year = date.year
    month = date.month
    yearly_data_power[year][month].append(power_usage_mw)

f_power.close()

# 월별 평균 전력 사용량 계산
average_usage_power = []
for month in range(1, 13):
    monthly_values = []
    for year in range(2020, 2025):
        if year == 2024 and month > 8:  # 2024년은 8월까지만 포함
            continue
        if yearly_data_power[year][month]:
            monthly_values.append(np.mean(yearly_data_power[year][month]))
    average_usage_power.append(np.mean(monthly_values) if monthly_values else 0)

# 미세먼지 데이터 처리
f_pm = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data_pm = csv.reader(f_pm)

header_pm = next(data_pm)

# 월별 데이터를 저장할 딕셔너리 생성
monthly_data_pm = {month: [] for month in range(1, 13)}

# 데이터 처리
for row in data_pm:
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    if 2019 <= date.year <= 2024:
        month = date.month
        monthly_data_pm[month].append(float(row[1]))

f_pm.close()

# 월별 평균 미세먼지 값 계산
monthly_avg_pm = [np.mean(monthly_data_pm[month]) if monthly_data_pm[month] else 0 for month in range(1, 13)]

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 Y축에 전력 사용량 그래프 그리기
line_power, = ax1.plot(range(1, 13), average_usage_power, marker='o', label='Average Power Usage (MW)', color='b')
ax1.set_xlabel('Month')
ax1.set_ylabel('Power Usage (MW)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 두 번째 Y축 생성
ax2 = ax1.twinx()

# 두 번째 Y축에 미세먼지 그래프 그리기
line_pm, = ax2.plot(range(1, 13), monthly_avg_pm, marker='o', label='Avg PM', color='r')
ax2.set_ylabel('PM', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# X축 레이블 설정
ax1.set_xticks(range(1, 13))
ax1.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 그래프 제목
plt.title('Average Monthly Power Usage (MW) and PM Data (2020-2024)')

# 범례 추가
ax1.legend(loc='upper left', fontsize=10)  # 전력 사용량에 대한 범례
ax2.legend(loc='upper right', fontsize=10)  # 미세먼지에 대한 범례

# 레이아웃 자동 조정
fig.tight_layout()

# 그래프 출력
plt.show()
```

- 월별 미세먼지 통합

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 파일 열기
f = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별 월별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {month: [] for month in range(1, 13)} for year in range(2019, 2025)}

# 데이터 처리
for row in data:
    # 첫 번째 열 (날짜)을 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 2019년부터 2024년까지의 데이터만 필터링, 2024년은 11월까지만
    if 2019 <= date.year <= 2024:
        if date.year == 2024 and date.month > 11:
            continue  # 2024년 12월 이후는 제외
        # 해당 연도와 월에 맞는 미세먼지 값 추가
        month = date.month
        yearly_data[date.year][month].append(float(row[1]))

# 파일 닫기
f.close()

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 연도별 평균 미세먼지 값 계산 후 그래프 그리기
for year in range(2019, 2025):
    # 각 월별 평균 미세먼지 값 계산 (2024년은 11월까지만)
    monthly_avg_pm = [np.mean(yearly_data[year][month]) if yearly_data[year][month] else 0 for month in range(1, 13)]
    
    # 2024년은 11월까지만 표시
    if year == 2024:
        monthly_avg_pm = monthly_avg_pm[:11]
    
    # 그래프에 그리기
    plt.plot(range(1, len(monthly_avg_pm) + 1), monthly_avg_pm, marker='o', label=f'{year}')

# 그래프 제목과 레이블 설정
plt.title('Average Monthly PM Data by Year (2019-2024)')
plt.xlabel('Month')
plt.ylabel('PM')
plt.grid(True)
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 레이아웃 자동 조정
plt.tight_layout()
plt.legend(title='Year')

# 그래프 출력
plt.show()
```

- 전력사용량

```python
import csv
import datetime
import matplotlib.pyplot as plt

# 파일 열기
f = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data = csv.reader(f)

# 헤더를 읽고 건너뛰기
header = next(data)

# 연도별, 월별 데이터를 저장할 딕셔너리 생성
yearly_data = {year: {month: [] for month in range(1, 13)} for year in range(2020, 2025)}

# 데이터 처리
for row in data:
    # 쉼표를 제거하고 전력 사용량을 float으로 변환 후, MW로 변환
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    
    # 날짜를 datetime 객체로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    # 연도와 월에 따라 전력 사용량을 저장
    year = date.year
    month = date.month
    yearly_data[year][month].append(power_usage_mw)

# 파일 닫기
f.close()

# 월별 평균 전력 사용량 계산 (연도별로)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# 그래프 그리기
plt.figure(figsize=(10, 6))

# 각 연도별 데이터 그리기
colors = ['b', 'g', 'r', 'c', 'm']  # 각 연도별로 사용할 색상 리스트
for i, year in enumerate(range(2020, 2025)):
    # 2024년의 경우 8월까지만 표시
    if year == 2024:
        months_to_plot = range(1, 9)  # 2024년은 1월부터 8월까지
    else:
        months_to_plot = range(1, 13)  # 다른 연도는 1월부터 12월까지
    
    # 각 월별 평균 전력 사용량 계산
    avg_monthly_usage = [
        sum(yearly_data[year][month]) / len(yearly_data[year][month]) if len(yearly_data[year][month]) > 0 else 0 
        for month in months_to_plot
    ]
    
    # 연도별 그래프 그리기
    plt.plot([months[month - 1] for month in months_to_plot], avg_monthly_usage, label=f'{year}', color=colors[i], marker='o')

# 그래프 제목과 레이블 설정
plt.title('Monthly Power Usage (MW) from 2020 to 2024 (2024 only until Aug)')
plt.xlabel('Month')
plt.ylabel('Power Usage (MW)')
plt.grid(True)
plt.legend(title='Year')

# X축 레이블 회전
plt.xticks(rotation=45)

# 레이아웃 자동 조정
plt.tight_layout()

# 그래프 출력
plt.show()
```

- 평균 사용전력과 미세먼지양 비교

```python
import csv
import datetime
import matplotlib.pyplot as plt
import numpy as np

# 전력사용량 데이터 처리
f_power = open('전력사용량 2020~2024.csv', encoding='UTF-8')
data_power = csv.reader(f_power)

header_power = next(data_power)

# 연도별, 월별 데이터를 저장할 딕셔너리 생성
yearly_data_power = {year: {month: [] for month in range(1, 13)} for year in range(2020, 2025)}

# 데이터 처리
for row in data_power:
    power_usage_watt = float(row[1].replace(',', ''))
    power_usage_mw = power_usage_watt / 1_000_000  # W에서 MW로 변환
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    year = date.year
    month = date.month
    yearly_data_power[year][month].append(power_usage_mw)

f_power.close()

# 월별 평균 전력 사용량 계산
average_usage_power = []
for month in range(1, 13):
    monthly_values = []
    for year in range(2020, 2025):
        if year == 2024 and month > 8:  # 2024년은 8월까지만 포함
            continue
        if yearly_data_power[year][month]:
            monthly_values.append(np.mean(yearly_data_power[year][month]))
    average_usage_power.append(np.mean(monthly_values) if monthly_values else 0)

# 미세먼지 데이터 처리
f_pm = open('미세먼지 2019~2024.csv', encoding='UTF-8-sig')
data_pm = csv.reader(f_pm)

header_pm = next(data_pm)

# 월별 데이터를 저장할 딕셔너리 생성
monthly_data_pm = {month: [] for month in range(1, 13)}

# 데이터 처리
for row in data_pm:
    date = datetime.datetime.strptime(row[0], '%b-%y')
    
    if 2019 <= date.year <= 2024:
        month = date.month
        monthly_data_pm[month].append(float(row[1]))

f_pm.close()

# 월별 평균 미세먼지 값 계산
monthly_avg_pm = [np.mean(monthly_data_pm[month]) if monthly_data_pm[month] else 0 for month in range(1, 13)]

# 그래프 그리기
fig, ax1 = plt.subplots(figsize=(10, 6))

# 첫 번째 Y축에 전력 사용량 그래프 그리기
line_power, = ax1.plot(range(1, 13), average_usage_power, marker='o', label='Average Power Usage (MW)', color='b')
ax1.set_xlabel('Month')
ax1.set_ylabel('Power Usage (MW)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# 두 번째 Y축 생성
ax2 = ax1.twinx()

# 두 번째 Y축에 미세먼지 그래프 그리기
line_pm, = ax2.plot(range(1, 13), monthly_avg_pm, marker='o', label='Avg PM', color='r')
ax2.set_ylabel('PM', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# X축 레이블 설정
ax1.set_xticks(range(1, 13))
ax1.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

# 그래프 제목
plt.title('Average Monthly Power Usage (MW) and PM Data (2020-2024)')

# 범례 추가
ax1.legend(loc='upper left', fontsize=10)  # 전력 사용량에 대한 범례
ax2.legend(loc='upper right', fontsize=10)  # 미세먼지에 대한 범례

# 레이아웃 자동 조정
fig.tight_layout()

# 그래프 출력
plt.show()
```
